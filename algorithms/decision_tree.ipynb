# 决策树

---

决策树的生成是一个递归过程，通常有三种情形导致递归返回：1）当前结点包含的样本全属于同一类别，无需划分；2）当前属性集合为空，或是所有样本在所有属性上取值相同，无法划分；3）当前结点包含的样本集合为空，不能划分。

一般而言，随着划分过程的不断进行，我们希望决策树的分支结点所包含的样本尽可能属于同一个类别，即结点的“纯度-purity”越来越高。因此，划分属性的选择倾向于获得“纯度提升”大的属性。

决策树所形成的分类边界的特点是轴平行，由多个与坐标轴平行的分段组成。

多变量决策树（multivariate decision tree）可以实现更复杂的划分边界，如“斜划分”；其学习过程不再是为每个非叶结点寻找一个最优划分属性，而是试图在属性的线性组合上建立一个合适的分类器。
