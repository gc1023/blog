# ZB - 面试总结

---

面试是一个双方相互了解的机会，作为应聘者，没有必要刻意降低自己的身位；抓住一切机会去了解自己正在申请的岗位和公司的情况，以便于自己做出一个更为明智、理性的决定。


## 面试快结束的时候，可以向招聘人员询问哪些信息？

1. 公司、部门介绍
2. 面试岗位的主要工作内容
3. 请招聘人员对自己做一个评价


## 2018-08-23 | 加推科技一面
---

面试官对我的评价：
**1、机器学习基础知识不扎实**
2、思路清晰
3、项目经验上缺乏亮眼的成绩

项目介绍：几乎每个项目都问了一下，评价是这些项目与算法的相关性都不是很强；重点介绍了JDDC比赛、招标信息采集与挖掘、宝安区城市运行指数项目、TMA。

知识点：
1、LSTM三个门结构的原理
2、SVM常用核函数
3、决策树的三个划分属性选择标准的数学公式
4、维特比算法解决CRF的预测问题
5、bagging、boosting、模型融合
6、RNN中的步长、层数
7、2014 science发表文章中的聚类算法
8、卡方检验
9、LDA算法

问题：
1、异常值检测的常用方法
2、特征筛选的常用方法
3、如何判断一个数据集是否满足正态分布，不满足的话，怎样进行转换


反思：
这个面试非常深入，历时约1个半小时。面试官是大数据架构师，两年前开始了解机器学习方面的知识。整个面试过程中，项目方面的内容都描述的还行，算法思想和实现流程方面也答的还行，但是算法方面更为细节性的 知识，如计算公式等，表现非常差。
